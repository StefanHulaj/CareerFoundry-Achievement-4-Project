Step 5

```python
ords_prods_merge_new['max_order'] = ords_prods_merge_new.groupby(['user_id'])['order_number'].transform(np.max)
```

Adding flags for loyal, new and regular customers.

```python
ords_prods_merge_new.loc[ords_prods_merge_new['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge_new.loc[(ords_prods_merge_new['max_order'] <= 40) & (ords_prods_merge_new['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge_new.loc[ords_prods_merge_new['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge_new['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge_new.groupby('loyalty_flag')['prices'].mean()
```

There is some noticable differences in prices. Loyal customers tend to purchase less expensive items then new and regular customers.

Step 6

```python
ords_prods_merge_new['avg_spending_user'] = ords_prods_merge_new.groupby(['user_id'])['order_number'].transform(np.max)
```

Adding flags for high and low spenders

```python
ords_prods_merge_new.loc[ords_prods_merge_new['avg_spending_user'] > 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge_new.loc[ords_prods_merge_new['avg_spending_user'] <= 10, 'spending_flag'] = 'High spender'
```

```python
#Confirming results
ords_prods_merge_new[['user_id', 'avg_spending_user', 'spending_flag']].head(20)
```

Step 7

```python
ords_prods_merge_new['ords_freq_median'] = ords_prods_merge_new.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

Adding flags for frequent, non-frequent and regular customers

```python
ords_prods_merge_new.loc[ords_prods_merge_new['ords_freq_median'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge_new.loc[(ords_prods_merge_new['ords_freq_median'] <= 20) & (ords_prods_merge_new['ords_freq_median'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge_new.loc[ords_prods_merge_new['ords_freq_median'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
#Confirming results
ords_prods_merge_new[['user_id', 'ords_freq_median', 'frequency_flag']].head(20)
```
